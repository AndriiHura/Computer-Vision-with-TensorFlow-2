```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import ndimage
```

```python
image = plt.imread("segmentation.jpeg")
print(image.shape)
plt.imshow(image)
```

# Change image to 1 channel grayscale

```python
gray = rgb2gray(image)
plt.imshow(gray, cmap='gray')
```

```python
gray.shape
```

```python
gray_r = gray.reshape(gray.shape[0]*gray.shape[1])
for i in range(gray_r.shape[0]):
    if gray_r[i] > gray_r.mean():
        gray_r[i] = 1
    else: 
        gray_r[i] = 0

gray = gray_r.reshape(gray.shape[0], gray.shape[1])
plt.imshow(gray, cmap='gray')
```

#  Let's do multiple thresholds:

```python
gray = rgb2gray(image)
gray_r = gray.reshape(gray.shape[0]*gray.shape[1])

for i in range(gray_r.shape[0]):
    if gray_r[i] > gray_r.mean():
        gray_r[i] = 3
    elif gray_r[i] > 0.5:
        gray_r[i] = 2
    elif gray_r[i] > 0.25:
        gray_r[i] = 1
    else:
        gray_r[i] = 0

gray = gray_r.reshape(gray.shape[0], gray.shape[1])
plt.imshow(gray, cmap='gray')
```

# Edge Detection Segmentation
For this reason we will use two filters, which are called sobel filters:
Horizontal

And Vertical

```python
image = plt.imread('index.png')
plt.imshow(image)
```

```python
image.shape
```

```python
gray = rgb2gray(image)

#Defining sobel filter matrices
sobel_horizontal = np.array([[1,2,1], [0,0,0], [-1,-2,-1]])
print(sobel_horizontal, ' is a filter for detecting horizontal edges')

sobel_vertical = np.array([[-1,0,1], [-2,0,2], [-1,0,1]])
print(sobel_vertical, ' is a filter for detecting vertical edges')
```

## Now we will convolve this filter over our image using ndimage from scipy

```python
out_h = ndimage.convolve(gray, sobel_horizontal, mode='reflect')
out_v = ndimage.convolve(gray, sobel_vertical, mode='reflect')

# here mode determines how the input array is extended when the filter overlaps a border.
```

```python
plt.imshow(out_h, cmap="gray")
```

```python
plt.imshow(out_v, cmap="gray")
```

### It is possible to detect both horizontal and vertical edges at once using 1 filter. This filter is called laplace filter


```python
kernel_laplace = np.array([[1,1,1], [1,-8,1], [1,1,1]])
print(kernel_laplace, 'is a laplacian kernel')
```

### Now let's try to convolve it over our image

```python
out_l = ndimage.convolve(gray, kernel_laplace, mode='reflect')
plt.imshow(out_l, cmap='gray')
```

### Just for experiment sake let's try apply it to first landscape image

```python
image1 = plt.imread("segmentation.jpeg")
gray1 = rgb2gray(image1)
plt.imshow(gray1, cmap='gray')
```

```python
out_l1 = ndimage.convolve(gray1, kernel_laplace, mode='reflect')
plt.imshow(out_l1, cmap='gray')
```
